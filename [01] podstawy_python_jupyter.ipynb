```python
#tworzenie i wyświetlanie tablicy
studenci = ["Ania", "Zosia", "Kasia", "Franek"]

print(studenci)
```

```python
#warianty użycia funkcji if w zależności od ilości elementów w tablicy
for student in studenci:
    print("student ma na imie:", student)
    
for student in studenci:
    print(student)
    if student[0]=="A":
        print("student na poczatku alfabetu")

for numer in range(10):
    print(numer)
```

```python
#funkcja enumerate w pythonie, wyswietla indeksy i nazwy w tablicy
for numer, student in enumerate(studenci):
    print("student {} ma numer {}".format(student,numer))
```

```python
#wyswietlanie danych z dwóch różnych tabel, na dwa różne sposoby
#funkcja zip, która łączy dwie tabele w jedną

oceny = [2,2,5,3]

for numer, student in enumerate(studenci):
    print("student {} ma numer {} i ocenę {}".format(student,numer,oceny[numer]))

print("--"*20)    
for number, (student, ocena) in enumerate(zip(studenci,oceny)):
    print("student {} ma numer {} i ocenę {}".format(student,number,ocena))
```

```python
#### challenge 0.2
zajecia = [["wf"],
           ["informatyka", "nauka czytania"],
           ["relax", "jedzenie"],
           ["mispy","mispy w domu"],
           ["piateczek", "praca domowa z mispow", "Ilovemispy", "mispy w czasie snu"]
          ]
dnityg= ["Pon","Wto","Sro","Czwar","Pio"]

# zdefiniuj funkcje ktora poda ile jest zajec danego dnia
#z wykorzystaniem funkcji enumerate - zmienna "daydzien" przechowuje nazwe dnia, zmienna "liczbadni" przechowuje indeks dnia, którego obliczana jest liczba elementów za pomocą funkcji "len" 
def zlicz(zajecia,dnityg):
  for liczbadni, daydzien in enumerate(dnityg):
    print("Liczba zajęć w {} : {}".format(daydzien, len(zajecia[liczbadni])))
zlicz(zajecia,dnityg)
print('---'*8)

# lista_zajec_z_dnia -> ile ich jest
#podobna funkcja do powyzszej, ale dodatkowo podaje zajecia w danym dniu "zajecia[liczbadni]"
def zlicz2(zajecia,dnityg):
  for liczbadni, daydzien in enumerate(dnityg):
    print("Zajęcia w {} (liczba zajęć {}) : {} ".format(daydzien, len(zajecia[liczbadni]), zajecia[liczbadni]))
zlicz2(zajecia,dnityg)
print('---'*8)

# po
# posortuj dni po najbardziej zajetych
# petla for zlicza liczbe zajec w danym dniu, następnie łączy (zipuje) liczbę zajęć w dniu, nazwe dnia i zajęcia z dnia w jedną zmienna, na koniec lista jest sortowana malejąco i wyswietlane są po najbardziej zajetych
liczbazajecc = [0]*5
for dziennumer,dziendzien in enumerate(dnityg):
  liczbazajecc[dziennumer] = len(zajecia[dziennumer])
zaj = list(zip(liczbazajecc,dnityg,zajecia))
zaj = sorted(zaj, reverse=True)
print(zaj)

print('---'*8)
#### challenge 0.3
# napisz program ktory zdajdzie dzien z najwieksza iloscia 
# baza funkcji taka sama jak powyżej, sprawdzana jest największa liczba zajęć z wykorzystaniem funkcji "max"
temp1 = [0]*5
def zlicz3(zajecia,dnityg):
  for liczbadni, daydzien in enumerate(dnityg):
    temp1[liczbadni] = len(zajecia[liczbadni]) 
  print("Dzien z najwieksza iloscia zajec ({}) : {}".format(max(temp1), daydzien,))
zlicz3(zajecia,dnityg)
print('---'*8)
```
