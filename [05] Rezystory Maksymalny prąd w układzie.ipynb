```
from pulp import *
problem = LpProblem("Maksymalny prad",LpMaximize)

x1 = LpVariable("i1", 0, None, LpInteger)
x2 = LpVariable("i2", 0, None, LpInteger)
x3 = LpVariable("i3", 0, None, LpInteger)
x4 = LpVariable("i4", 0, None, LpInteger)
x5 = LpVariable("i5", 0, None, LpInteger)
R1 = 8
R2 = 6
R3 = 4
R4 = 10
R5 = 8

problem += x1 + x2 + x3 + x4 + x5, "Maksymalny prad w ukladzie"
problem += x1 + x2 <= 4, "suma pierwszego węzła"
problem += x4 + x5 <= 4, "suma drugiego węzła"
problem += (x1*R1)/R2 <= 3, "rezystor 1"
problem += (x2*R2)/R1 <= 2, "rezystor 2"
problem += x3 <= 4, "rezystor 3"
problem += (x4*R4)/R5 <= 2, "rezystor 4"
problem += (x5*R5)/R4 <= 2, "rezystor 5"

problem.writeLP("pradproblem.lp")
problem.solve()
print("Status:", LpStatus[problem.status])
```

```
for v in problem.variables():
    print (v.name, "=", v.varValue)
print ("Maksymalny prad w ukladzie: {} A".format(value(problem.objective)))
```
